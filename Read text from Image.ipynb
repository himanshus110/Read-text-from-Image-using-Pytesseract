```python
import pytesseract
import cv2 
import numpy as np
```

```python
def image_to_text(image_path):
  image = cv2.imread(image_path)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  gray = cv2.GaussianBlur(gray, (7,7), 0)
  text = pytesseract.image_to_string(gray)
  print(text)
```

```python
image_path_1="/content/1.jpg"
image_to_text(image_path_1)
```

```python
image_path_2="/content/2.jpg"
image_to_text(image_path_2)
```

```python
image_path_3="/content/3.jpg"
image_to_text(image_path_3)
```
